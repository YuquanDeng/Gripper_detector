```python



# Load JSON data
import json
    
json_path = "/home/nil/manipulation/dataset/pick_apple_100_328/line/pick_apple_0/train/episode_0_point.json"

with open(json_path, 'r') as f:
    data = json.load(f)

traj = []
for sample in data:
    # img = Image.open(sample['image'])
    # traj = sample['line']
    # print(sample["point"])
    traj.append(sample['point'])

print("length of traj: ", len(traj))
    
```

```python
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
%matplotlib inline

img = Image.open(data[0]['image'])

draw = ImageDraw.Draw(img)
for i in range(len(traj) - 1):
    draw.line((traj[i][0], traj[i][1], traj[i + 1][0], traj[i + 1][1]), fill="red", width=5)

# Display the image in the notebook
plt.imshow(img)
plt.axis('off')  # Optional: Hide axes for better visualization
plt.show()
```
